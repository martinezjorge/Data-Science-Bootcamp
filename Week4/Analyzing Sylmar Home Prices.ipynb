## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
KNN = KNeighborsClassifier(n_neighbors=4)
xtrain,xtest,ytrain,ytest = train_test_split(XY,Y)
model = KNN.fit(xtrain,ytrain)
ypredict = model.predict(xtest)
accuracy_score(ytest,ypredict)
```

```python
confusion_matrix(ytest,ypredict)
```

```python
def evaluate_model(X,Y, method, repeats):
    accs=[]
    for repeat in range(repeats):
        xtrain,xtest,ytrain,ytest = train_test_split(X,Y)
        method.fit(X,Y)
        YP=method.predict(xtest)
        accs.append(accuracy_score(ytest,YP))
    return (accs)
```

```python
stats4 = evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=4),25)
stats5 = evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=5),25)
stats6 = evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=6),25)
```

```python
plt.boxplot([stats4,stats5,stats6]);
```

## Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GNB_stats = evaluate_model(XY,Y,GaussianNB(),25)
```

```python
plt.boxplot([stats4,stats5,stats6,GNB_stats]);
plt.xticks([1,2,3,4],['KNN-4','KNN-5','KNN-6','GNB']);
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
def show_classifications(ax,X,Y,classifier,nsteps):
    
    classifier.fit(X,Y)

    colors = ["red","blue","green","purple"]
    markers = ["o","s","v","d"]
    
    x=X.T[0]
    y=X.T[1]

    classvals=np.unique(Y)
    nclasses=len(classvals)

    levels=np.arange(0.5,nclasses,1)

    for i ,classval in enumerate(classvals):
        xpts,ypts = (X[Y==classval]).T
        ax.scatter(xpts,ypts,c=colors[i],marker=markers[i])


    # generate mesh of xy points that nsteps x nsteps in size
    xmesh,ymesh = np.meshgrid(
        np.linspace(x.min(),x.max(),nsteps),
        np.linspace(y.min(),y.max(),nsteps))


    # unwind the array into a single long one-dimensional array
    unwound_mesh = np.c_[
        xmesh.ravel(),
        ymesh.ravel()]


    unwound_prediction = classifier.predict(unwound_mesh)
    prediction = unwound_prediction.reshape(xmesh.shape)

    # Generate a contour plot
    ax.contourf(xmesh,ymesh,prediction, alpha=0.1)

    # Draw thick black lines between the contour levels
    ax.contour(xmesh,ymesh,prediction, alpha=1,levels=levels,colors='k')
    ax.set_xlim(x.min(),x.max())
    ax.set_ylim(y.min(),y.max())

    return ax
```

```python
fig,ax = plt.subplots()
KNN = KNeighborsClassifier(n_neighbors=4)
show_classifications(ax,XY,Y,KNN,150)
ax.set_xlabel("Area", fontsize=16)
ax.set_ylabel("Price/Sq Foot", fontsize=16)
ax.grid()
ax.set_title("KNN",fontsize=18)
fig.set_size_inches(10,6)
```

```python
fig,ax = plt.subplots()
GNB = GaussianNB()
show_classifications(ax,XY,Y,GNB,150)
ax.set_xlabel("Area", fontsize=16)
ax.set_ylabel("Price/Sq Foot", fontsize=16)
ax.grid()
ax.set_title("GNB",fontsize=18)
fig.set_size_inches(10,6)
```

```python
fig,ax = plt.subplots()
LDA = LinearDiscriminantAnalysis()
show_classifications(ax,XY,Y,LDA,150)
ax.set_xlabel("Area", fontsize=16)
ax.set_ylabel("Price/Sq Foot", fontsize=16)
ax.grid()
ax.set_title("LDA",fontsize=18)
fig.set_size_inches(10,6)
```
