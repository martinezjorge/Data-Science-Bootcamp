# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
KNN=KNeighborsClassifier(n_neighbors=3)
xtrain,xtest,ytrain,ytest=train_test_split(XY,Y)
model=KNN.fit(xtrain,ytrain)
ypredict=model.predict(xtest)
accuracy_score(ytest,ypredict)
```

```python
confusion_matrix(ytest,ypredict)
```

```python
# write a function that runs a model many times and gives me the statistics on the accuracy so I can plot 
# a box plot
```

```python
def evaluate_model(X, Y, method, repeats):
    accs=[]
    for repeat in range(repeats):
        xtrain,xtest,ytrain,ytest=train_test_split(X,Y)
        method.fit(X,Y)
        YP=method.predict(xtest)
        accs.append(accuracy_score(ytest,YP))
    return(accs)
```

```python
stats3=evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=3),25)
stats4=evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=4),25)
stats5=evaluate_model(XY,Y,KNeighborsClassifier(n_neighbors=5),25)
```

```python
plt.boxplot([stats3,stats4,stats5]);
```

```python
## Naive Bayes
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
GNB=GaussianNB()
xtrain,xtest,ytrain,ytest=train_test_split(XY,Y)
model=GNB.fit(xtrain,ytrain)
YP=model.predict(xtest)
accuracy_score(ytest,YP)
```

```python
GNBSTATS=evaluate_model(XY,Y,GNB,25)
```

```python
plt.boxplot([stats3,stats4,stats5, GNBSTATS]);
plt.xticks([1,2,3,4],["KNN-3","KNN-4","KNN-5","GNB"]);
```

```python
# LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
LDA = LinearDiscriminantAnalysis()
model=LDA.fit(xtrain,ytrain)
YP=model.predict(xtest)
accuracy_score(ytest,YP)
```

```python
LDASTATS=evaluate_model(XY,Y,LDA,25)
```

```python
plt.boxplot([stats3,stats4,stats5, GNBSTATS, LDASTATS]);
plt.xticks([1,2,3,4,5],["KNN-3","KNN-4","KNN-5","GNB", "LDA"]);
```

```python
# Quadratic Discriminant
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
QDA = QuadraticDiscriminantAnalysis()
model=QDA.fit(xtrain,ytrain)
YP=model.predict(xtest)
accuracy_score(ytest,YP)
```

```python
QDASTATS=evaluate_model(XY,Y,QDA,25)
```

```python
plt.boxplot([stats3,stats4,stats5, GNBSTATS, LDASTATS, QDASTATS]);
plt.xticks([1,2,3,4,5, 6],["KNN-3","KNN-4","KNN-5","GNB", "LDA", "QDA"]);
```
