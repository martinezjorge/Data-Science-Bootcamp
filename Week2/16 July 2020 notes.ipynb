# Matplotlib.pyplot library

```python
import matplotlib.pyplot as plt
```

Two interfaces to the library:

1. Direct pyplot interface, easiest to use if all commands for a single plot are in a single cell, difficult if you want to refer to a figure that has been created earlier
2. Object oriented interface -- create a "plot" object in the background with various plot features, refer to the object with a variable name, and make changes to it at any point you want to in any program

```python
x1=(1,1)
x2=(2,4)
x3=(3,2)
x4=(4,3)
```

pyplot wants all the x-values in one array and all the y values in another array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdata=np.array([1,2,3,4])
ydata=np.array([1,4,2,3])
```

If the numbers come as pairs, you could convert them using zip and splat

```python
X,Y=list(zip(*[x1,x2,x3,x4]))
X=np.array(X)
Y=np.array(Y)
print(X)
print(Y)
```

What is a scatter plot? A bunch of dots, on an x-y coordinates

```python
A=np.array([1,1.5,2.5,4])
B=np.array([1.5,2,3,3.5])
```

Using the Pyplot interface

```python
plt.plot(X,Y, marker="o", markerfacecolor="red",
        linestyle="--", color="green",
        label="Physics Majors");
plt.plot(A,B, linestyle="-.", label="Biology Majors")
plt.xlabel("Number of Sick Days", fontsize=18);
plt.ylabel("Times I went to the Pub", fontsize=18)
plt.title("Data Science and the World",fontsize=20)
plt.legend(fontsize=14);
plt.grid()
plt.gcf().set_size_inches(10,6)
```

Using the Object Oriented Interface

Every Plot has exactly one fig (figure) and at least one or more Axes

```python
fig, ax=plt.subplots() # ax is an Axes object that I just created
ax.plot(X,Y, marker="o", color="red", label="first line")
ax.set_xlabel("Pubs",fontsize=16)
ax.set_ylabel("Inns",fontsize=16)
```

```python
ax.plot(A,B,marker="v",color="green",label="whiffle balls")
ax.legend(fontsize=14)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2,sharey=True)
fig.set_size_inches(10,4)
axs[0].plot(X,Y,color="purple")
axs[1].plot(A,B,color="green")
axs[0].set_title("this")
axs[1].set_title("that",fontsize=18)
axs[1].grid()
axs[0].set_xlabel("breakfast")
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2,sharey=True)
fig.set_size_inches(10,4)
axs[0][0].plot(X,Y,color="purple")
axs[0][0].plot(A,B,color="blue")

axs[1][1].plot(A,B,color="green")
axs[0][0].set_title("top left",fontsize=18)
axs[1][1].set_title("bottom right",fontsize=18)
fig.set_size_inches(9,9)
```

```python
xdata = np.linspace(0, 4*np.pi, 250)
```

```python
# define a workaround to tangent for plotting
# this workaround edits out points near the asymptote

mytan=np.tan(xdata)
mytan[:-1][np.diff(mytan)<0]=np.nan
```

```python
fig,ax=plt.subplots()
ax.plot(xdata, np.cos(xdata), label="cosine")
ax.plot(xdata, np.sin(xdata), label="sine", lw=3, ls=":")
ax.plot(xdata, mytan, label="tan", lw=3, ls="--")
#ax.plot(xdata, np.tan(xdata))
ax.set_ylim([-1,1])
# lw = line weight
# ls = line style
ax.legend(fontsize=14)
fig.set_size_inches(10,6)
ax.grid()
```

```python
from scipy.stats import norm # normal distribution 
xdata = np.linspace(200, 800, 100)
y=norm(500,100).pdf(xdata) # pdf = prob density function applied to xdata
plt.plot(xdata, y)
```

```python
first,second,theplot=plt.hist(lottastuff,
        bins=np.linspace(200,800,13),                      
        color="hotpink",edgecolor="blue",
        density=1, label="simulated")
plt.plot(xdata, y, label="theoretical")
plt.grid()
plt.legend(fontsize=14)
plt.xlabel("SAT Score", fontsize=16)
plt.ylabel("Relative Frequency", fontsize=16)
plt.gcf().set_size_inches(10,6)
fig = plt.gcf()
fig.savefig("MyPrettyPicture.png",dpi=150)
```
