```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
samplesize = 10
```

```python
np.random.seed(123)
np.random.normal(500,100, samplesize)
```

To really get integers:

```python
first_sample = np.round(np.random.normal(500,100, samplesize)).astype(int)
second_sample = np.round(np.random.normal(500,100, samplesize)).astype(int)
```

```python
plt.boxplot([first_sample,second_sample]);
plt.grid()
```

Five Number Summary: Minimum, 25%, 50%, 75%, Maximum

```python
samples = [np.round(np.random.normal(500,100, samplesize)).astype(int) for sample in range(20)]
```

```python
print(samples)
```

Boxplots are good for comparing distributions and samples

Don't always assume that the median is where the mean is.

```python
plt.boxplot(samples);
plt.grid()
plt.gcf().set_size_inches(12,6)
plt.title("Sample of Size 10 from N(500,100)", fontsize=16)
```

95% of the sample means to fall where:

$ \dfrac{2\sigma}{\sqrt{n}}$ of 500

The SD of the sample means should be:

$\dfrac{\sigma}{\sqrt{n}}$

```python
s = 2*100 / np.sqrt(10) # 95 %
```

```python
sample_means = [np.mean(s) for s in samples]
print(sample_means)
```

```python
plt.boxplot(samples);
plt.grid()
plt.gcf().set_size_inches(12,6)
plt.title("Sample of Size 10 from N(500,100)", fontsize=16)

plt.axhline(y=500+s, xmin=0,xmax=21)
plt.axhline(y=500-s,xmin=0,xmax=21)

# plt.axvline()

plt.scatter(range(1,len(sample_means)+1),sample_means)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
homes = pd.read_csv("homes-91325.csv")
homes
```

```python
homes.columns
```

```python
SF = np.array(homes["SF"])
print(SF)
```

```python
homes["PRICE"]
```

```python
# Convert to a number
```

```python
np.array(homes["PRICE"])
```

```python
Y = [float(p.strip('$').replace(',',"")) for p in homes['PRICE']]
print(Y)
```

```python
plt.scatter(SF,Y)
plt.xlabel("Square Feet", fontsize=14)
plt.ylabel("Price", fontsize=14)
plt.grid()
```

```python
xbar = np.mean(SF)
ybar = np.mean(Y)
sx = np.std(SF, ddof=1)    # ddof=1 means divide by n-1 in standard deviation, not n
sy = np.std(Y,ddof=1)
r=np.corrcoef(SF,Y)[0,1]
b = r*sy/sx
a = ybar - (b*xbar)
print(f"a={a}, b={b}")
```

```python
from scipy.stats import linregress
```

```python
linregress(SF,Y)
```

```python
plt.scatter(SF,Y)
plt.xlabel('Sqare Feet')
plt.ylabel('Price')
plt.grid()
plt.plot([1000,6500],[a+b*1000, a+b*6500])
```
