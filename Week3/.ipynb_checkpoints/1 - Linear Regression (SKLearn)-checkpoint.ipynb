```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('AnscombesQuartet.csv')
```

```python
data
```

```python
x1 = data.x1.to_numpy()
y1 = data.y1.to_numpy()

x2 = data.x2.to_numpy()
y2 = data.y2.to_numpy()

x3 = data.x3.to_numpy()
y3 = data.y3.to_numpy()

x4 = data.x4.to_numpy()
y4 = data.y4.to_numpy()
```

```python
model = LinearRegression()
x1 = x1.reshape(-1,1)
model.fit(x1,y1)
Y1 = model.predict(x1)
print(model.coef_)
print(model.intercept_)
```

```python
model = LinearRegression()
x2 = x2.reshape(-1,1)
model.fit(x2,y2)
Y2 = model.predict(x2)
print(model.coef_)
print(model.intercept_)
```

```python
model = LinearRegression()
x3 = x3.reshape(-1,1)
model.fit(x3,y3)
Y3 = model.predict(x3)
print(model.coef_)
print(model.intercept_)
```

```python
model = LinearRegression()
x4 = x4.reshape(-1,1)
model.fit(x4,y4)
Y4 = model.predict(x4)
print(model.coef_)
print(model.intercept_)
```

```python
fig, ax = plt.subplots(2,2)

fig.suptitle("Anscombe's Quartet", fontsize=20)

# Dataset 1
ax[0][0].set_title("Dataset 1")
ax[0][0].set_xlabel("$X_1$")
ax[0][0].set_ylabel("$Y_1$")

ax[0][0].scatter(x1, y1, color='purple', label='Data Points')
ax[0][0].plot(x1, Y1, color='black', label='Regression')
ax[0][0].legend()

# Dataset 2
ax[0][1].set_title("Dataset 2")
ax[0][1].set_xlabel("$X_2$")
ax[0][1].set_ylabel("$Y_2$")

ax[0][1].scatter(x2,y2, color='purple', label='Data Points')
ax[0][1].plot(x2, Y2, color='black', label='Regression')
ax[0][1].legend()

# Dataset 3
ax[1][0].set_title("Dataset 3")
ax[1][0].set_xlabel("$X_3$")
ax[1][0].set_ylabel("$Y_3$")

ax[1][0].scatter(x3, y3, color='purple', label='Data Points')
ax[1][0].plot(x3, Y3, color='black', label='Regression')
ax[1][0].legend()

# Dataset 4
ax[1][1].set_title("Dataset 4")
ax[1][1].set_xlabel("$X_4$")
ax[1][1].set_ylabel("$Y_4$")

ax[1][1].scatter(x4, y4, color='purple', label='Data Points')
ax[1][1].plot(x4, Y4, color='black', label='Regression')
ax[1][1].legend()

# Fig stuff
fig.set_size_inches(10,10)
fig.tight_layout()
fig.subplots_adjust(top=0.9)
plt.savefig("Anscombe's Quartet.png")
```
