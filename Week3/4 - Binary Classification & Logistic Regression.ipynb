```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import label_binarize
from sklearn.metrics import RocCurveDisplay, roc_curve, roc_auc_score, plot_roc_curve
from sklearn.metrics import ConfusionMatrixDisplay,  plot_confusion_matrix
from sklearn.metrics import accuracy_score, auc
```

```python
data = pd.read_csv("COVID19_line_list_data.csv")
```

```python
data = data[['age','death']].dropna()
```

```python
age = data.age.to_numpy().astype('int')
death = data.death.tolist()
```

```python
binary_death = []
for d in death:
    if d == '0':
        binary_death.append(0)
    else:
        binary_death.append(1)
death = np.array(binary_death)
```

```python
death = label_binarize(death, classes=[0, 1])
```

```python
age_train, age_test, death_train, death_test = train_test_split(age, death)
```

```python
age_train = age_train.reshape(-1,1)
age_test  = age_test.reshape(-1,1)
death_test = death_test.reshape(-1,1)
```

```python
model = LogisticRegression()
stuff = model.fit(age_train, death_train)

death_preds = model.predict(age_test)
probs = model.predict_proba(age_test)
acc = accuracy_score(death_test, death_preds)
```

```python
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(2):
    fpr[i], tpr[i], _ = roc_curve( death_test, probs[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
# plt.plot(fpr[1], tpr[1], color="navy",
#          label=f"Logistic Regression (AUC={roc_auc[1]:.3f})")
```

```python
print(f"Accuracy: {acc}")

ConfusionMatrixDisplay(plot_confusion_matrix(model, death_test, death_preds))
plt.show()
```

# All Data (No Train Test Split)

```python
plot_roc_curve(model, age.reshape(-1,1), death)

plt.title("COVID-19 1000 Cases: Age. vs Survival\nEntire Dataset", fontsize=18)
plt.xlabel("True Positive Rate", fontsize=14)
plt.ylabel("False Positive Rate", fontsize=14)

plt.plot([0,1],[0,1], color='black', lw=2, linestyle="--")
plt.legend(loc='upper left')
plt.gcf().set_size_inches(10,10)
```

# 10 Train Test Splits

```python
fig, ax = plt.subplots()
for _ in range(10):
    age_train, age_test, death_train, death_test = train_test_split(age, death)
    plot_roc_curve(model, age_test.reshape(-1,1), death_test, ax=ax)

ax.set_title("COVID-19 1000 Cases: Age. vs Survival \n10 Train/Test Splits", fontsize=16)
ax.set_xlabel("False Positive Rate", fontsize=14)
ax.set_ylabel("True Positive Rate", fontsize=14)
ax.plot([0,1],[0,1], color='black', lw=2, linestyle="--")
ax.legend(loc='lower right')
fig.set_size_inches(10,10)
fig.show()
```
