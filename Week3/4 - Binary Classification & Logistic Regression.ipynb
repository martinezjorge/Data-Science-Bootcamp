```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("COVID19_line_list_data.csv")
```

```python
data = data[['age','death']].dropna()
```

```python
data
```

```python
age = data.age.to_numpy().astype('int')
death = data.death.tolist()
```

```python
print(len(age))
print(len(death))
```

```python
binary_death = []
for d in death:
    if d == '0':
        binary_death.append(0)
    else:
        binary_death.append(1)
death = np.array(binary_death)
```

```python
print(len(death))
```

```python
age_train, age_test, death_train, death_test = train_test_split(age, death)
```

```python
age_train = age_train.reshape(-1,1)
age_test  = age_test.reshape(-1,1)
# death_train = death_train.reshape(-1,1)
death_test = death_test.reshape(-1,1)
```

```python
print(len(death_train))
print(len(age_train))
```

```python
model = LogisticRegression()
model.fit(age_train,death_train)
death_preds = model.predict(age_test)
ConfusionMatrixDisplay(plot_confusion_matrix(model, death_test, death_preds));
```
